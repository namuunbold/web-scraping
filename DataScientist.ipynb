```python
import requests
from bs4 import BeautifulSoup
import time

url = 'https://www.simplyhired.com/search?q=data+scientist'
job_urls = []
count = 0

while count < 300:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract cursor value from the "data-nxt-cursor" attribute of the "a" element with class "chakra-link css-gxlopd"
    next_page_link = soup.find('a', {'class': 'chakra-link css-gxlopd'})
    if next_page_link:
        cursor = next_page_link.get('data-nxt-cursor')
    else:
        next_page_link = soup.find('a', {'aria-label': 'next page'})
        if next_page_link:
            cursor = next_page_link.get('href').split('&cursor=')[1]
        else:
            print("No next page available.")
            break

    # Append the cursor value to the URL as a query string parameter named "cursor"
    url = f'{url}&cursor={cursor}'

    # Scrape the job URLs from the current page
    job_list = soup.find("ul", {"id": "job-list"})
    if job_list:
        for job in job_list.find_all("li"):
            job_url = job.a.get('href')
            job_urls.append(job_url)
            count += 1
            if count >= 300:
                break

    # Sleep for 1 second to avoid hitting the server too frequently
    time.sleep(1)

full_job_url = []
for i, job_url in enumerate(job_urls):
    temp_url = "https://www.simplyhired.com" + job_url
    full_job_url.append(temp_url)
    print(f"Job {i+1} URL: {temp_url}")
```

```python
import pymongo 
from pymongo import MongoClient
from bs4 import BeautifulSoup
client = MongoClient('localhost', 27017)
db = client['SimplyHired']

for i in range(1, 301):
    file = open(f"/Users/namuunboldbaatar/Downloads/SimplyHired/data-scientist/SH_DataScientist_{i}.htm", 'r', encoding="utf-8")
    data = file.read()
    file.close()
    soup = BeautifulSoup(data, "html.parser")
    
    title = soup.find("h2", {"data-testid": "viewJobTitle"})
    if title:
        title = title.text.strip()
    else:
        title = ''

    job_details = soup.find_all("span", {"class": "chakra-stack css-xtodu4"})

    company_elem = soup.find("span", {"data-testid": "viewJobCompanyName"})
    company = company_elem.find("span", {"data-testid": "detailText"}).text.strip() if company_elem else ''

    salary_elem = soup.find("span", {"data-testid": "viewJobBodyJobCompensation"})
    salary = salary_elem.find("span", {"data-testid": "detailText"}).text.strip() if salary_elem else ''

    posted_elem = soup.find("span", {"data-testid": "viewJobBodyJobPostingTimestamp"})
    posted = posted_elem.find("span", {"data-testid": "detailText"}).text.strip() if posted_elem else ''

    location_elem = soup.find("span", {"data-testid": "viewJobCompanyLocation"})
    location = location_elem.find("span", {"data-testid": "detailText"}).text.strip() if location_elem else ''

    job_type_elem = soup.find("span", {"data-testid": "viewJobBodyJobDetailsJobType"})
    job_type = job_type_elem.find("span", {"data-testid": "detailText"}).text.strip() if job_type_elem else ''

    company_rating_elem = soup.find("span", {"data-testid": "viewJobCompanyRating"})
    company_rating = company_rating_elem.text.strip() if company_rating_elem else ''
    if company_rating:
        company_rating = company_rating.strip().split()[0]
    else:
        company_rating = ''

    description_elem = soup.select_one('div[data-testid="viewJobBodyJobFullDescriptionContent"]')
    description = description_elem.text.strip() if description_elem else ''

    data = {
        "title": title,
        "company": company,
        "location": location,
        "job_type": job_type,
        "company_rating": company_rating,
        "salary": salary,
        "posted": posted,
        "description": description
    }

    db.jobs.insert_one(data)
```
